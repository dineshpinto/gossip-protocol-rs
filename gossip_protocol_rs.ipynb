```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

import seaborn as sns

from gossip_protocol_rs import run_gossip_protocol

sns.set_theme()
```

## Variation with number of neighbours

```python
df = pd.DataFrame()

for num_peers in range(3, 10):
    res = run_gossip_protocol(
        num_honest_sample=5,
        num_adversarial_sample=4,
        num_non_sample=1000,
        num_peers=num_peers,
        cycles=100
    )
    df[f"K={num_peers}"] = res
```

```python
df = df.replace([np.inf, -np.inf], np.nan)
```

```python
fig, ax = plt.subplots()

sns.lineplot(data=df, ax=ax)

ax.legend()
ax.set_title("Variation with number of neighbors K")
ax.set_xlabel("Cycles")
ax.set_ylabel("Convergence")
```
